```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Sample data
documents = [
    "C:/Users/HP/OneDrive/Desktop/Internship/xlsheets/alice.txt"
]

# Initialize CountVectorizer
tf_vectorizer = CountVectorizer()

# Fit and transform the data
tf = tf_vectorizer.fit_transform(documents)

# Create DataFrame
df = pd.DataFrame(tf.toarray(), columns=tf_vectorizer.get_feature_names_out())

# Export to CSV
df.to_csv('alice_cv_2.csv', index=False)

print("CSV file 'alice_cv_2.csv' has been created successfully.")
```
